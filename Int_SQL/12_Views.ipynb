### Create the VIEW
#### daily_revenue

```python
%%sql

CREATE VIEW daily_revenue AS
SELECT
    orderdate,
    SUM(quantity * netprice * exchangerate) AS total_revenue
FROM sales
GROUP BY
    orderdate
```

### View the VIEW in Query

```python
%%sql

SELECT *
FROM daily_revenue
```

### Remove the VIEW
#### Delete with SQL Query

```python
%%sql

DROP VIEW daily_revenue;
```
